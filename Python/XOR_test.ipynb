```python
import tensorflow as tf
import keras as kr
import numpy as np
import datetime

# Build the model
model = kr.Sequential()
model.add(kr.layers.Flatten(input_shape=(2,)))
model.add(kr.layers.Dense(4, activation='relu'))
model.add(kr.layers.Dense(4, activation='relu'))
model.add(kr.layers.Dense(4, activation='relu'))
model.add(kr.layers.Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss="mse", optimizer=kr.optimizers.Adam(learning_rate=0.01), metrics=['accuracy'])

# Create a TensorBoard callback
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

# tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir="./logs")

# Input data
state = np.array([[0,0],
                  [0,1],
                  [1,0],
                  [1,1]])
label = np.array([0,1,1,0])

# Train the model with the TensorBoard callback
model.fit(state, label, epochs=150, verbose=1, callbacks=[tensorboard_callback])
```

```python
test = np.array([[0,0],
                  [0,1],
                  [1,0],
                  [1,1]])
model.predict(test)
```
